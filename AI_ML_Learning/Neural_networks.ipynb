```python
#NEURAL NETWORKS - RNN
import numpy as np

# Imports Sequential model used to build neural networks layer by layer
from tensorflow.keras.models import Sequential

# Embedding  -> Converts words/numbers into dense vector representations
# SimpleRNN -> Processes sequential data and remembers previous information
# Dense      -> Fully connected layer used for prediction/output
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense

# Tokenizer converts text words into numerical tokens (integers)
from tensorflow.keras.preprocessing.text import Tokenizer

# Converts class labels into one-hot encoded format
from tensorflow.keras.utils import to_categorical

# Makes all sequences equal length by adding padding values
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
sentence = "I love deep learning"
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([sentence])
```

```python
sequence = tokenizer.texts_to_sequences([sentence])[0]
```

```python
sequence
```

```python
x = [] #stores input sequences
y = [] #stores expected outputs
for i in range(1, len(sequence)):
  x.append(sequence[:i])
  y.append(sequence[i])

print("x : ", x)
print()
print("y : ", y)
```

```python
x = pad_sequences(x)
print(x)
```

```python
x = np.array(x)
y = np.array(y)
print(x)
print("\n", y)
```

```python
y = to_categorical(y)
y
```

```python
# Calculates total vocabulary size by counting all unique words
# +1 is added because index 0 is reserved for padding
vocab_size = len(tokenizer.word_index) + 1
```

```python
# Model creation
model = Sequential()
# Creates an empty neural network model where layers are added one after another

# Embedding layer
model.add(Embedding(input_dim=vocab_size, output_dim=10, input_length=x.shape[1]))
# Converts input word indices into dense vectors
# input_dim = total vocabulary size
# output_dim = each word is represented using a vector of size 10
# input_length = length of each input sequence

# RNN layer
model.add(SimpleRNN(20))
# Adds a Simple Recurrent Neural Network layer with 20 neurons
# This layer processes sequential data and remembers previous information

# Dense layer
model.add(Dense(vocab_size, activation='softmax'))
# Output layer with number of neurons equal to vocabulary size
# softmax activation gives probability for predicting the next word
```

```python
#compile
#Optimizer → learns by updating weights
#Loss Function → measures mistakes
#Accuracy → measures performance
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Trains the neural network using input data (x) and output data (y)
# epochs=500 -> the model learns from the entire dataset 500 times
# verbose=0 -> hides training progress/output during training
model.fit(x, y, epochs = 500, verbose = 0)

test_text = "I love deep"
test_seq = tokenizer.texts_to_sequences([test_text])
test_seq
```

```python
test_seq = pad_sequences(test_seq, maxlen=x.shape[1])
test_seq
```

```python
prediction = model.predict(test_seq)
```

```python
prediction
```

```python
predicted_index = np.argmax(prediction)
# Finds the index of the highest value in the prediction array
# The index represents the predicted word/class

predicted_index
# Displays the predicted index value
```

```python
#LSTM - Long Shot-term Memory : used to store information for long period of time
#since rnn forgets previous information and causes vanishing gradient problem , lstm is used which is an advanced rnn
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.sequence import pad_sequences

sentence = "I love pizza"

tokenizer = Tokenizer()

tokenizer.fit_on_texts([sentence])

sequence = tokenizer.texts_to_sequences([sentence])[0]
print("sequence : ")
print(sequence)
print()

x = []
y = []
for i in range(1, len(sequence)):
    x.append(sequence[:i])
    y.append(sequence[i])

print("x : ", x)
print("y : ", y)
print()

print("x after padding : ")
x = pad_sequences(x)
print(x)
print()

x = np.array(x)
y = np.array(y)

print("After performing to_categorical on output sequence y  : ")
y = to_categorical(y)
print(y)
print()

vocab_size = len(tokenizer.word_index) + 1
print("vocab size : ", vocab_size)

model = Sequential()
#Embedding layer
model.add(Embedding(input_dim=vocab_size, output_dim=10, input_length=x.shape[1]))
#LSTM layer
model.add(LSTM(20))
#Dense layer
model.add(Dense(vocab_size, activation='softmax'))

#compile
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.fit(x, y, epochs=500, verbose=0)

test_text = "I love"
test_seq = tokenizer.texts_to_sequences([test_text])
test_seq = pad_sequences(test_seq, maxlen=x.shape[1])
print("Test sequence : ", test_seq)
print()

prediction = model.predict(test_seq)
print("prediction : ", prediction)
print()
predicted_index = np.argmax(prediction)
print("Predicted Index : ", predicted_index)

for word, index in tokenizer.word_index.items():
    if index == predicted_index:
        print("Predicted Word : ", word)
        break

```

```python
#Attention Mechanism
# Imports NumPy library for arrays and numerical operations
import numpy as np
# Imports TensorFlow library for tensors and deep learning operations
import tensorflow as tf
```

```python
# Creates a list of words in the sentence
words = ['I', 'love', 'pizza']
```

```python
#embeddings for the words I, love, pizza
embeddings = np.array([[1, 0],
                       [0, 2],
                       [1, 2]])
# Converts words into numerical vector representations
# I     -> [1,0]
# love -> [0,2]
# pizza -> [1,2]
```

```python
# Prints the embedding matrix
print(embeddings)
```

```python
#create query
# Creates the Query vector
# Query represents what we are searching/focusing on
Q = tf.constant([[1.0, 2.0]])

# Prints the Query tensor
print(Q)
```

```python
#create keys
# Converts embeddings into Key tensor
# Keys are used to calculate relevance with Query
k = tf.constant(embeddings, dtype = tf.float32)

# Prints the Key tensor
print(k)
```

```python
#create values
# Converts embeddings into Value tensor
# Values contain actual information/content
v = tf.constant(embeddings, dtype = tf.float32)

# Prints the Value tensor
print(v)
```

```python
# Performs matrix multiplication between Query and transposed Keys
# Calculates attention scores
# transpose_b=True transposes the Key matrix before multiplication
scores = tf.matmul(Q,k, transpose_b = True)

# Prints attention scores for each word
print("scores : ", scores)
print()

# Applies softmax activation to attention scores
# Converts scores into probability values
# Higher score gets higher attention weight
# All weights together sum to 1
weights = tf.nn.softmax(scores)

# Prints attention weights for each word
print("Attention weights :", weights)
print()

# Multiplies attention weights with Value matrix
# Produces the final attention output
# Important words contribute more to the output
output = tf.matmul(weights, v)

# Prints the final attention output tensor
print("Output : ", output)
```

```python
#Multi head attention
import tensorflow as tf
x = tf.constant([[
    [1.0, 0.0],
    [0.0, 2.0],
    [1.0, 2.0]
]], dtype = tf.float32)
print("x shape : ",x.shape)
print()

# Creates a Multi-Head Attention layer
# num_heads=2 -> uses 2 separate attention heads
# key_dim=2 -> size/dimension of Query and Key vectors for each head
mha = tf.keras.layers.MultiHeadAttention(num_heads=2, key_dim=2)

# Applies multi-head attention
# First x -> Query (Q)
# Second x -> Key (K)
# Third x -> Value (V)
# Since all are x, this is called Self-Attention
output = mha(x, x, x)

# Prints the final output after multi-head attention
print(output)
```
