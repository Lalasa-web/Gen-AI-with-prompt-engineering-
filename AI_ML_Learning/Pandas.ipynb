```python
#data cleaning
#step 1 : check number of null values
df.isnull().sum()
```

```python
df.dropna() #removes all the rows with null values, does not affect the original dataframe as we have not did inplace = True
```

```python
df
```

```python
df.dropna(how = "any") #if any row that had null value
```

```python
df.dropna(how = "all") # if all the values in the row had null values then drop that row
```

```python
df
```

```python
df.fillna(0)
```

```python
df["Age"].fillna(df["Age"].mean()) #we have filled the null value with mean in the age column
```

```python
df["Salary"].fillna(df["Salary"].median()) #we have filled the null value with median in the salary column
```

```python
df
```

```python
df["Age"].fillna(df["Age"].mean())
```

```python
#forward fill
df["Age"].fillna(method = "ffill") # replaces the missing value with the value above to it
```

```python
#backward fill
df["Age"].fillna(method = "bfill") # replaces the missing value with the value above to it
```

```python
df["Name"] = df["Name"].replace("Charlie", "Rose")
df
```

```python
#Duplicates
df_dup = df[df.duplicated()]
df_dup
```

```python
df_dup = df[df.duplicated(keep = "first")] #"it keeps first element and marks the one that is duplicated"
df_dup
```

```python
df_dup = df[df.duplicated(keep = "last")] #"it keeps last element and marks the one that is duplicated"
df_dup
```

```python
df = df.drop_duplicates()
df
```

```python
#invalid data
#lambda
df["Promoted Salary"] = df["Salary"] * 10
df
```

```python
#invalid data
#lambda
df["Promoted Salary"] = df["Promoted Salary"].apply(lambda x: x/10 if x > 650000 else x)
df
```

```python
#apply and lamda
#without using lambda
def multiplying_age(x):
  return x * 2

df["Age"] = df["Age"].apply(multiplying_age)
```

```python
df
```

```python
df["Age"] = df["Age"].apply(lambda x:x/2) #with lambda usage
```

```python
df
```

```python
#joins and merges
department_info = {
    "Dept" : ["HR", "IT", "Finance"],
    "Location" : ["New York", "San Francisco", "Chicago"],
    "Manager" : ["Laura", "Steve", "Naina"]
}

df2 = pd.DataFrame(department_info)
```

```python
df2
```

```python
#concat
pd.concat([df, df2]) #the two datasets have been stacked horizanally
```

```python
pd.concat([df, df2], axis = 1) #the two datasets have been stacked vertically
```

```python
# we merge the two datasets to not to have the same column repeated
pd.merge(df, df2, on = "Dept")
```
