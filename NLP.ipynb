```python
#performing cleaning, bag of words, TF-IDF, cbow and skpi-gram
#step1 -> import libraries
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

from gensim.models import Word2Vec

#dowload required nltk resources
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')

#step 2: original texts

text1 = "Virat kohli scored an amazing century yesterday 🏏🔥"
text2 = "I hate waiting in long traffic jams every morning 😡🚙"
text3 = "Open AI launched a powerful new AI model in california 🚀🤖"
text4 = "The pizza was extremely delicious 😋🍕!!!"
text5 = "Amazon delivered my package very late 😤📦 #badservice"
text6 = "Dhoni loves playing cricket with young players 🏏❤️"
text7 = "Microsoft announced new cloud services in New York🗽☁️"

documents = [text1, text2, text3, text4, text5, text6, text7]

#step 3: initialize stop words and lemmatizer
stop_words = (stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

#step 4: text cleaning
cleaned_documents = []
for text in documents:
    #convert to lower case
    text = text.lower()
    #remove special characters
    text = re.sub(r'[^a-zA-Z\s]','',text)
    #tokenization
    tokens = word_tokenize(text)
    #remove stop words
    tokens = [word for word in tokens if word not in stop_words]
    #lemmatization
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    #join tokens back into sentence
    cleaned_text = " ".join(tokens)
    cleaned_documents.append(cleaned_text)

#print cleaned text
print("\n\nCleaned documents: ")
for doc in cleaned_documents:
    print(doc)

#step 5:

#Bag of words
print("Bag of words: ")
bow = CountVectorizer()
bow_result = bow.fit_transform(cleaned_documents)
print("\nFeatures:\n")
print(bow.get_feature_names_out())
print("\nBag of words matrix: \n")
print(bow_result.toarray())
#TF-IDF
print("TF-IDF: ")
tfidf = TfidfVectorizer()
tfidf_result = tfidf.fit_transform(cleaned_documents)
print("\nFeatures: \n")
print(tfidf.get_feature_names_out())
print("\nTF-IDF matrix: \n")
print(tfidf_result.toarray())

#step 6:

#prepare data for Word2Vec
tokenized_sentences = [doc.split() for doc in cleaned_documents]
print("\nTokenized sentences : \n")
print(tokenized_sentences)
#CBOW model
print("\nCBOW model: ")
cbow_model = Word2Vec(tokenized_sentences, vector_size = 5, window = 2, min_count = 1, sg = 0)
print("\nVector representation of 'cricket': \n")
print(cbow_model.wv['cricket'])
#skip-gram model
print("\nSkip-gram model: ")
skipgram_model = Word2Vec(tokenized_sentences, vector_size = 5, window = 2, min_count = 1, sg = 1)
print("\nVector representation of 'cricket': \n")
print(skipgram_model.wv['cricket'])

```
