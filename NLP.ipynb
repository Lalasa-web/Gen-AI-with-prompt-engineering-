```python
#nlp -> natural language processing
#import regular expressions re
#import natural language toolkit
import re
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt') #download punkt tokenizer
nltk.download('punkt_tab') #helper data for tokenizer
nltk.download('stopwords') #downloads a collection of stopwords (ex : is, am, the, and, of, an)
nltk.download('wordnet') #Downloads the WordNet dictionary database.
```

```python
text = "Wow!!! the new i-phone costs $999.99 in 2026 😯🤯 can you believe it???"
```

```python
text.lower()
```

```python
# remove special characters
text = re.sub(r'[^a-zA-Z\s]','',text)
```

```python
text
```

```python
#tokenization
tokens = word_tokenize(text)
tokens
```

```python
#stopwords
# Get English stopwords

stop_words = set(stopwords.words('english'))
stop_words
```

```python
#filtered data after removing stop words
tokens = [token for token in tokens if token not in stop_words]
tokens
```

```python
#lemmatization -> converts original word into root/base word
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
tokens = [lemmatizer.lemmatize(token) for token in tokens] # costs -> cost
tokens
```

```python
#Bag of words (BOW)
from sklearn.feature_extraction.text import CountVectorizer
```

```python
sentences = ["I ate pizza in the morning",
              "I ate rotis"]
```

```python
#create object
cv = CountVectorizer()
```

```python
#learn vocabulary + convert text
#fit -> learns all unique words
#transform -> converts text into vectors
bow = cv.fit_transform(sentences)
print(cv.get_feature_names_out())
print(bow.toarray())
```

```python
#TF - IDF
from sklearn.feature_extraction.text import TfidfVectorizer
documents = ['I am playing cricket',
             "He played cricket yesterday"]
tfidf = TfidfVectorizer()
result = tfidf.fit_transform(documents)
print("features : ")
print(tfidf.get_feature_names_out())
print(result.toarray())
```

```python
from gensim.models import Word2Vec
```

```python
sentences = [["Rohit", "likes", "eating", "mangoes", "daily"],
             ["Virat", "likes", "playing", "cricket"],
             ["kids", "love", "eating", "chocolates"]]
```

```python
#sg = 0 means cbow
#sg = 1 means skip-gram
model = Word2Vec(sentences, vector_size = 5, window = 2, min_count = 1, sg = 0)
print(model.wv["mangoes"])
```

```python
#sg = 0 means cbow
#sg = 1 means skip-gram
model = Word2Vec(sentences, vector_size = 5, window = 2, min_count = 1, sg = 1)
print(model.wv["mangoes"])
```
