```python
#data cleaning
#step 1 : check number of null values
df.isnull().sum()
```

```python
df.dropna() #removes all the rows with null values, does not affect the original dataframe as we have not did inplace = True
```

```python
df
```

```python
df.dropna(how = "any") #if any row that had null value
```

```python
df.dropna(how = "all") # if all the values in the row had null values then drop that row
```

```python
df
```

```python
df.fillna(0)
```

```python
df["Salary"].fillna(df["Salary"].median()) #we have filled the null value with median in the salary column
```

```python
df # here I have run the dataframe creation cell() to get the age column back and ran "df" in the current cell
```

```python
df["Age"].fillna(df["Age"].mean())
```

```python
#forward fill
df["Age"].fillna(method = "ffill") # replaces the missing value with the value above to it
```

```python
#backward fill
df["Age"].fillna(method = "bfill") # replaces the missing value with the value above to it
```

```python
df["Name"] = df["Name"].replace("Charlie", "Rose")
df
```

```python
#Duplicates
df_dup = df[df.duplicated()]
df_dup
```

```python
df_dup = df[df.duplicated(keep = "first")] #"it keeps first element and marks the one that is duplicated"
df_dup
```

```python
df_dup = df[df.duplicated(keep = "last")] #"it keeps last element and marks the one that is duplicated"
df_dup
```

```python
df = df.drop_duplicates()
df
```

```python
#invalid data
#lambda
df["Promoted Salary"] = df["Salary"] * 10
df
```

```python
#invalid data
#lambda
df["Promoted Salary"] = df["Promoted Salary"].apply(lambda x: x/10 if x > 650000 else x)
df
```
